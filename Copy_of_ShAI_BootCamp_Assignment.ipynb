```python
range = df.describe().loc['max'] - df.describe().loc['min']
range
```

```python
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
#filling the null values with mean is the best way instead of dropping them
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Notes', 'Status'], axis=1) #removed object datatypes
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['OvertimePay', 'OtherPay'])  #only 4 null values in each, removing them won't cause a problem
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=5, color='blue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.show()
```

```python
department_proportions = df['JobTitle'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(department_proportions, labels=department_proportions.index)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
grouped_data = df.groupby('JobTitle')['TotalPay'].agg(['mean', 'median'])
correlation = df['TotalPay'].corr(df['BasePay'])
```

```python
grouped_data
```

```python
correlation
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['BasePay'], df['TotalPay'])
plt.title('Scatter Plot: BasePay vs TotalPay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```
